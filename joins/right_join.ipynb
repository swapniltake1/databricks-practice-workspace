```python
%sql
SELECT
    c.customer_id,
    c.full_name,
    a.account_id
FROM bankof420.bank_master.customers c
RIGHT JOIN bankof420.bank_core.accounts a
    ON c.customer_id = a.customer_id;
```

```python

customers = spark.table("bankof420.bank_master.customers")
accounts  = spark.table("bankof420.bank_core.accounts")


df_right = customers.join(
    accounts,
    "customer_id",
    "right"
)

df_right.display()
```

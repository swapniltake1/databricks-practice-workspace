### with sql

```python
%sql
USE CATALOG bankof420;
SHOW SCHEMAS;
SHOW TABLES IN bank_master;
```

```python
%sql
SELECT
    c.customer_id,
    c.full_name,
    a.account_id,
    a.account_type,
    a.balance
FROM bankof420.bank_master.customers c
INNER JOIN bankof420.bank_core.accounts a
    ON c.customer_id = a.customer_id;
```

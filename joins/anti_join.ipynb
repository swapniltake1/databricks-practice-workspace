```python
%sql
SELECT
    c.customer_id,
    c.full_name
FROM bankof420.bank_master.customers c
LEFT ANTI JOIN bankof420.bank_core.accounts a
    ON c.customer_id = a.customer_id;

```

```python
customers = spark.table("bankof420.bank_master.customers")
accounts  = spark.table("bankof420.bank_core.accounts")

result = customers.join(accounts, on="customer_id", how="left_anti").select("customer_id", "full_name")
display(result)
```

```python
df_anti = customers.join(
    accounts,
    "customer_id",
    "left_anti"
)

df_anti.display()
```

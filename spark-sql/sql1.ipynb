```python
df = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/databricks-datasets/samples/population-vs-price/data_geo.csv")
display(df)
```

```python
df.createOrReplaceTempView("geo_data")

result = spark.sql(
    """
    SELECT City, `2014 Population estimate`
    FROM geo_data
    WHERE `2014 Population estimate` > 1000000
    ORDER BY `2014 Population estimate` DESC
    """
)
display(result)
```

```python
%sql
-- Count the number of cities in the dataset
SELECT COUNT(DISTINCT City) AS num_cities
FROM geo_data
```

```python
%sql
-- Find the average population estimate
SELECT AVG(`2014 Population estimate`) AS avg_population
FROM geo_data
```

```python
%sql
-- Group by State and sum the population estimates
SELECT State, SUM(`2014 Population estimate`) AS total_population
FROM geo_data
GROUP BY State
ORDER BY total_population DESC
```

```python
%sql
-- Filter cities in California with population over 500,000
SELECT City, State, `2014 Population estimate`
FROM geo_data
WHERE State = 'California' AND `2014 Population estimate` > 500000
ORDER BY `2014 Population estimate` DESC
```

```python
%sql
-- Get the top 5 most populous cities
SELECT City, State, `2014 Population estimate`
FROM geo_data
ORDER BY `2014 Population estimate` DESC
LIMIT 5
```

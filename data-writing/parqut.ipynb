```python
df = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/databricks-datasets/samples/population-vs-price/data_geo.csv")

df.display(5)
```

```python
parquet_path = "/tmp/parquet/data_geo"
```

```python
df.write \
    .format("parquet") \
    .mode("overwrite") \
    .save(parquet_path)
```

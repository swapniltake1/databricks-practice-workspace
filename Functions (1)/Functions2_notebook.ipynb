### READING DATA for next Functions

```python
df = spark.read.format('csv').option('inferSchema', 'true')\
    .option('header', 'true')\
    .load("/Volumes/raw-data/banking/csv/Banking_Database.csv")

df.display()
```

## Renaming calumn --> withColumnRenamed

```python
df.withColumnRenamed('Customer ID', 'customer_id').display()
```

## with column
- create new column 
- modify columns 

```python
from pyspark.sql.functions import lit 
df = df.withColumn('flag', lit("New"))
df.display()
```

```python
from pyspark.sql.functions import col
df.withColumn('multiply', col('Customer Id') * col('Age')).display()
```

### Modify the exsting column

```python
from pyspark.sql.functions import *
df = (
    df.withColumn(
        'Gender',
        regexp_replace('Gender', 'Female', 'F')
    )
    .withColumn(
        'Gender',
        regexp_replace('Gender', 'Male', 'M')
    )
)

display(df)
```

## Type Casting 

```python
df = df.withColumn('Age', col('Age').cast('int').cast(StringType()))

df.printSchema()
display(df)
```

### Sort / Order by 

```python
df.sort(col('Age').desc()).display()
```

```python
df.sort(col('Account Balance').asc()).display()
```

```python
df.sort(['Age', 'Account Balance'], ascending=[0, 1]).display()
```

### Limit Function


```python
df.limit(100).display()
```

### Drop



```python
df.drop('Age', 'Account Balance').display()
```

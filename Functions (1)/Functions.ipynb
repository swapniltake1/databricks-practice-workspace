## Loading data


```python
df = spark.read.format('csv').option('inferSchema', 'true')\
    .option('header', 'true')\
    .load("/Volumes/databricksdata/practice/csvdata/Comprehensive_Banking_Database.csv")

df.display()
```

## Select Function

```python
df.select('Customer ID', 'First Name', 'Last Name', 'City').display()
```

```python
# Function segrigations we use this standezidw way 
from pyspark.sql.functions import col
df.select(col('Customer ID'), col('First Name'), col('Last Name'), col('City')).display()
```

### ALIAS IN PYSPARK

```python
from pyspark.sql.functions import col
df.select(col('Customer ID').alias('id'), col('First Name').alias('name'), col('Email').alias('email'), col('City').alias('city'), col('Contact Number').alias('contact'), col('Age').alias('age'), col('Gender').alias('gender'), col('Account Balance').alias('balance'), col('Last Transaction Date').alias('last_transaction_date')).display()
```

### FILTER DATA

```python
df.filter(
    (col('City') == 'New York') &
    (col('Gender') == 'Female') & 
    (col('Transaction Type') == 'Withdrawal')).display()
```

```python
df_and = df.filter(
    (df["Loan Status"] == "Approved") &
    (df["Loan Amount"] > 30000)
)

display(df_and)
```

```python
df_or = df.filter(
    (df["Loan Type"] == "Personal") |
    (df["Loan Type"] == "Auto")
)

display(df_or)
```

```python
df_mix = df.filter(
    ((df["Loan Type"] == "Mortgage") | (df["Loan Type"] == "Personal")) &
    (df["Loan Amount"] > 20000) &
    (df["Loan Status"] == "Approved")
)

display(df_mix)
```

```python
df_isin = df.filter(
    (df["Loan Type"].isin("Mortgage", "Auto", "Personal")) &
    (df["Loan Status"] == "Approved")
)

display(df_isin)
```

```python
df_multi = df.filter(
    (df["Credit Limit"] > 5000) &
    (df["Card Type"] == "Visa") &
    (df["Loan Status"] == "Approved")
)

display(df_multi)
```

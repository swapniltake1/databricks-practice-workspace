```python
from pyspark.sql import functions as F
df = (
    spark.read.format("csv")
        .option("inferSchema", "true")
        .option("header", "true")
        .load("/Volumes/raw-data/banking/csv/Banking_Database.csv")
)

df.printSchema()
df.display(5)
```

```python
from pyspark.sql import functions as F

# Split the Email column by '@' and get the username part (index 0)
display(
    df.select(
        "Email",
        F.get(F.split("Email", "@"), 0).alias("Email_Username")
    )
)

# Split the Address column by spaces and get the first word (index 0)
display(
    df.select(
        "Address",
        F.get(F.split("Address", " "), 0).alias("Address_FirstWord")
    )
)

# Split the City column by '-' and get the second part (index 1), returns NULL if not present
display(
    df.select(
        "City",
        F.get(F.split("City", "-"), 1).alias("City_SecondPart")
    )
)
```

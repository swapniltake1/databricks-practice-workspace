### String Functions

```python
from pyspark.sql import functions as F
df = (
    spark.read.format("csv")
        .option("inferSchema", "true")
        .option("header", "true")
        .load("/Volumes/raw-data/banking/csv/Banking_Database.csv")
)

df.printSchema()
df.display(5)
```

```python
from pyspark.sql import functions as F

# Uppercase
display(
    df.select(
        "First Name",
        F.upper("First Name").alias("FirstName_Upper")
    )
)

# Lowercase
display(
    df.select(
        "Last Name",
        F.lower("Last Name").alias("LastName_Lower")
    )
)

# Initcap (capitalize first letter of each word)
display(
    df.select(
        "City",
        F.initcap("City").alias("City_InitCap")
    )
)

# Length of string
display(
    df.select(
        "Email",
        F.length("Email").alias("Email_Length")
    )
)

# Concatenate columns
display(
    df.select(
        "First Name",
        "Last Name",
        F.concat_ws(" ", "First Name", "Last Name").alias("Full_Name")
    )
)

# Substring
display(
    df.select(
        "Address",
        F.substring("Address", 1, 10).alias("Address_Substr")
    )
)

# Trim spaces
display(
    df.select(
        "Address",
        F.trim("Address").alias("Address_Trimmed")
    )
)

# Replace substring
display(
    df.select(
        "Email",
        F.regexp_replace("Email", "@gmail.com", "@example.com").alias("Email_Replaced")
    )
)
```
